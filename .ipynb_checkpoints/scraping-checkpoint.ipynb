```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# fetch all teams in euros 2024
```

```python
squads_url = "https://fbref.com/en/comps/676/schedule/European-Championship-Scores-and-Fixtures"
```

```python
squads_data = requests.get(squads_url)
```

```python
soup = BeautifulSoup(squads_data.text)
```

```python
squads_table = soup.select('table.stats_table')[0]
```

```python
links = squads_table.find_all('a')
links = [link.get('href') for link in links]
```

```python
squad_links = [link for link in links if '/squads' in link]
squad_links = list(set(squad_links))
squad_links
```

```python
# examplr url https://fbref.com/en/squads/c1e40422/Germany-Men-Stats
final_urls = [f"https://fbref.com{link}" for link in squad_links]
final_urls
```

```python
team_url = final_urls[0]
```

```python
data = requests.get(team_url)
```

```python
# to get shooting stats
data = requests.get(team_url)
soup = BeautifulSoup(data.text)
all_links_in_page = soup.find_all('a')
all_links_in_page = [link.get('href') for link in all_links_in_page if link.get('href') is not None]
all_links_in_page = [link for link in all_links_in_page if 'all_comps/shooting' in link]
all_links_in_page = list(set(all_links_in_page))
all_links_in_page
```

```python
shooting_url = f"https://fbref.com/" + all_links_in_page[0]
shooting_url
```

```python
data = requests.get(shooting_url)
shooting = pd.read_html(data.text, match='Shooting', flavor="bs4")[0]
shooting
```

```python
shooting.columns = shooting.columns.droplevel(0)
shooting.head()
```
