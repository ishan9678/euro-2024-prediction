```python
all_matches = []
```

```python
final_url = 'https://fbref.com/en/squads/7def9493/Romania-Men-Stats'

def add_year_param(url, year):
    url_parts = url.split('/')
    url_parts.insert(-1,year)
    modified_url = '/'.join(url_parts)
    return modified_url
    
print(add_year_param(final_url,'2022'))
```

```python
import requests
import pandas as pd
import time
import random
from bs4 import BeautifulSoup

for url in final_urls[8::]:

    years = ['2024','2022-2023','2022','2021','2020-2021','2018-2019','2018','2017','2016','2015','2014','2013','2012'
           ,'2010','2009','2008','2006','2004']
    team_name = url.split("/")[-1].replace('-Men-Stats', '')

    for year in years:
        scraping_url = add_year_param(url, year)

        print(scraping_url)

        data = requests.get(scraping_url)

        if '500 error' in data.text:
            print(f'error for {year}')
            continue
        matches_data = pd.read_html(data.text, match='Scores & Fixtures ')[0]

        # Fetch shooting stats
        soup = BeautifulSoup(data.text)
        all_links_in_page = [link.get('href') for link in soup.find_all('a') if link.get('href') is not None]
        shooting_links = [link for link in all_links_in_page if 'all_comps/shooting' in link]
        shooting_url = f"https://fbref.com{shooting_links[0]}"
        shooting_data = requests.get(shooting_url)
        shooting_stats = pd.read_html(shooting_data.text, match='Shooting')[0]
        shooting_stats.columns = shooting_stats.columns.droplevel(0)
        team_data = matches_data.merge(shooting_stats[['Date','Sh','SoT','SoT%','G/Sh','G/SoT','PK','PKatt']], on='Date')
        team_data['Year'] = year
        team_data['Team'] = team_name
        all_matches.append(team_data)
        print(f"{year} data added successfully")
        time.sleep(random.randint(10, 15))
    print(f'Data for {team_name} scraped successfully')
```

```python
combined_data = pd.concat(all_matches, ignore_index=True)
print(combined_data.tail())
```

```python
team_data['Date'] = pd.to_datetime(team_data['Date']).dt.strftime('%Y-%m-%d')
```

```python
combined_data.to_csv('all_matches.csv', encoding='utf-8', index=False)
```
